```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loaded_data = pd.read_csv('iris.csv')
print(loaded_data.head())
```

```python
print("Shape:", loaded_data.shape)
print("Columns:", loaded_data.columns)
print("\nInfo:")
print(loaded_data.info())
print("\nDescribe:")
print(loaded_data.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=loaded_data)
plt.title('Relationship b/w features(length and width)')
plt.show()
```

```python
loaded_data.hist(figsize=(10, 6))
plt.suptitle('Value Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=loaded_data,)
plt.title('Boxplot to identify outlers')

plt.show()
```
